```python
#####################################################
                    # TPA #
#####################################################

from pyspark import SparkContext
```

```python
sc = SparkContext()
```

```python
!hadoop fs -put /vagrant/TPA/CO2.csv
!hadoop fs -ls CO2.csv
```

```python
CO2_data = sc.textFile('CO2.csv')
```

```python
CO2_data.collect()
```

```python
#Making each model in one and only vector
#Première donnée : numéro de ligne - effacée
CO2_Separated_data = CO2_data.map(lambda x : x.split(','))
CO2_Separated_data.collect()
```

```python
#Structurer les données par colonnes
#La colonne 1 est la marque ou le modèle
#La colonne 2 est le Bonus/Malus
#La colonne 3 représente les rejets en C02
#La colonne 4 représente les coûts en énergie
CO2_Splitter = CO2_Separated_data.map(lambda x : ([value.strip() for value in x[1].split(',')], [value.strip() for value in x[2].split(',')], [value.strip() for value in x[3].split(',')], [value.strip() for value in x[4].split(',')]))

# Afficher le résultat
CO2_Splitter.collect()
```

```python
#################################################
#Nettoyage des données
#################################################

CO2_clean_rejet = CO2_Separated_data.map(lambda x: (x[0], x[1], [value.replace('\xa0', ' ').strip() for value in x[2].split(',')], x[3], x[4]))
CO2_clean_rejet.collect()
```

```python
##########################################
#Recuperation des donnees dans le fichier
#########################################

output_path = "./CO2.csv"
CO2_clean_rejet.saveAsTextFile(output_path)
```
